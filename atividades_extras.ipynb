# Exemplo 4.1

Uma memória RAM (MP) tem um espaço máximo de armazenamento de 2K. Cada célula pode armazenar 16 bits. Qual o valor total de bits que podem ser armazenados nesta memória e qual o valor de cada endereço?

```python
n = 2*2**10
m = 2**4

total = n * m
print(f"total = {total} bits")
print(f"total = {total/(2**10)}Kb = {total/((2**10)*8)}KB")
```

# Exemplo 4.3

Um computador, cuja memória RAM (MP) tem uma capacidade máxima de armazenamento de 2K palavras de 16 bits cada, possui um REM e um RDM. Qual o tamanho desses registradores; qual o valor do maior endereço dessa MP e qual a quantidade total de bits que nela podem ser armazenados?

```python
import math

# 2K = 2*(2**10) = 2**11 -> 11 bits de endereço
REM = math.log2(2*(2**10))
RDM = 2**4
total = 2*(2**10) * 2**4

print(f'''
REM = {int(REM)} bits
RDM = {RDM} bits
Maior endereço = 11111111111
Total armazenado = {total} bits = {int(total/2**10)}Kb = {int(total/(2**10*8))}KB
''')
```

Exemplo 4.4
Um processador possui um RDM com capacidade de armazenar 32 bits e um REM com capacidade de armazenar 24 bits. Sabendo-se que em cada acesso são lidas duas células da memória RAM (MP) e que o barramento de dados (BD) tem tamanho igual ao da palavra, pergunta-se:

a) Qual é a capacidade máxima de endereçamento do microcomputador em questão?

b) Qual é o total máximo de bits que podem ser armazenados na memória RAM (MP)?

c) Qual é o tamanho da palavra e de cada célula da máquina?

```python
import math

# REM = 24bits
# RDM = BD = 32 -> tamanho da célula = 32/2 = 16 bits
n = 2**24
m = 2**4
total = (n * m)/2**20
expoente = math.log2(total)


print(f'''
a) Capacidade máxima de endereçamento = 2^24 = {n}
b) Total máximo de bits armazenados = 2^{int(expoente)}Mb
c) Tamanho da palavra = BD = RDM = 32 bits; Tamanho da célula = {m}
''')
```
