# Questão 7
Descreva os barramentos que interligam o processador e MP, indicando função e direção do fluxo de sinais de cada um.

**Barramento de dados:** É um caminho para movimentação de dados no sistema. Por meio dele, uma informação pode ser adicionada ou recuperada da memória. O fluxo de dados é bidirecional, pois quando acionada a operação de escrita o sinal segue a direção Processador-Memória RAM, enquanto na operação de leitura a direção do sinal é Memória RAM-Processador.

**Barramento de endereços:** Designa a origem ou o destino dos dados no barramento de dados, indicando o local em que a informação será armazenada ou recuperada. A largura do BE determina a capacidade de memória máxima possível do sistema e suas linhas podem ser usadas para endereçar portas de E/S. O fluxo de dados é unidirecional, do processador para a memória RAM, pois o BE indica apenas qual local será acessado na RAM.

**Barramento de controle:** Controla o acesso e o uso das linhas de dados e endereço. Os sinais de controle transmitem informações de comando e temporização. Os sinais de temporização indicam a validade da informação de dados e endereço. Os sinais de comando especificam operações a serem realizadas. O fluxo de dados é bidirecional, podendo mandar sinais de leitura ou escrita do processador para a RAM, mas também de "Wait" no sentido inverso.

Você considera válida a afirmação "um computador com mais poder de processamento pode armazenar mais programas"?

Não, pois o poder de processamento se refere ao desempenho da CPU. O armazenamento permanente de programas é responsabilidade da memória secundária (HD/SSD). # Questão 34
Você considera válida a afirmação "vale aumentar a capacidade da memória principal para que o acesso aos meios magnéticos (discos rígidos e disquetes) seja mais rápido"?

Não. Do ponto de vista do hardware, aumentar a RAM não torna o acesso físico ao meio magnético mais rápido, visto que as propriedades físicas deste não serão alteradas, então a velocidade de acesso permanece a mesma. Contudo, o desempenho do sistema melhora, pois com a maior capacidade da RAM, menos requisições são feitas à memória secundária e a velocidade de resposta do sistema aumenta.
