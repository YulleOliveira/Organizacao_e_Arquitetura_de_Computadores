# Exemplo 5.2

Cálculo da quantidade de bits necessários para uma determinada memória cache.

Considere um sistema de computação com uma memória cache de 32KB de capacidade, constituída de linhas com 8 bytes de largura. A MP possui uma capacidade de 16MB.

```python
import math

total_dados = 2**5 * 2**10 * 8
largura_linha = 8 * 8
memoria_principal = 2**4 * 2**20 * 8

# Calculando o total de bits do campo da tag
q_linhas = total_dados/largura_linha
q_blocos = memoria_principal/largura_linha
largura_tag = math.log2(q_blocos/q_linhas)
total_tag = q_linhas * largura_tag

# total de bits da cache
memoria_cache = total_tag + total_dados

print(f'''Total dados = 32KB = 2**15B = {total_dados} bits
Total tag = {int(total_tag)} bits
Quantidade de bits necessária para a cache = {int(memoria_cache)} bits = 292KB''')
```

# Exemplo 5.3

Cálculo de formato de endereço para memórias cache com mapeamento direto.

Considere uma MP com 64MB de capacidade associada a uma memória cache que possui 2K linhas, cada uma com largura de 16 bytes. Determine o formato do endereço para ser interpretado pelo sistema de controle da cache.

```python
import math

memoria_principal = 2**6 * 2**20
q_linhas = 2**11
largura_linha = 16

# Quantidade de bits da linha
bits_linha = int(math.log2(q_linhas))
print(f"Linha: {bits_linha} bits")

# Quantidade de bits da tag
q_blocos = memoria_principal/largura_linha
blocos_por_linha = q_blocos/q_linhas
bits_tag = int(math.log2(blocos_por_linha))
print(f"Tag: {bits_tag} bits")

# Quantidade de bits da célula
total_bits = math.log2(memoria_principal)
bits_celula = int(total_bits - bits_tag - bits_linha)
print(f"Célula: {bits_celula} bits")

# Formato do endereço
print(f'''\nEndereço:
Tag         | Linha       | Célula
{"0" * bits_tag} | {"0" * bits_linha} | {"0" * bits_celula}''')
```

# Exemplo 5.4

Seja uma MP constituída de blocos com largura de 32 bytes, associada a uma cache com 128KB, em dado instante o processador realiza um acesso, colocando o seguinte endereço (expresso em algarismos hexadecimais): 3FC92B6. Determine qual deverá ser o valor binário da linha que será localizada pelo sistema de controle da cache.

```python
largura_bloco = 2**5
memoria_cache = 2**7 * 2**10
bits_endereco = 4 * 7
endereco = "3FC92B6"

# Quantidade de bits da célula
bits_celula = int(math.log2(largura_bloco))
print(f"Célula: {bits_celula} bits")

# Quantidade de bits da linha
q_linhas = memoria_cache/largura_bloco
bits_linha = int(math.log2(q_linhas))
print(f"Linha: {bits_linha} bits")

# Quantidade de bits da tag
bits_tag = bits_endereco - bits_celula - bits_linha
print(f"Tag: {bits_tag} bits")

# Endereço em binário
endereco_bin = "00" + str(bin(int(endereco, 16)))[2:29]
print(f"Endereço em binário: {endereco_bin}")

# Formato do endereço
print(f'''\nFormato do endereço:
Tag         | Linha        | Célula
{endereco_bin[0:11]} | {endereco_bin[11:23]} | {endereco_bin[23:29]}

Linha localizada pela cache: {endereco_bin[11:23]}''')
```

# Exemplo 5.5

Cálcula da quantidade de bits necessários para uma determinada memória cache.

Considere um sistema de computação com uma memória cache de 32KB de capacidade, constituída de linhas com 8 bytes de largura. A MP possui uma capacidade de 16MB.

```python
import math

total_dados = 2**5 * 2**10 * 8
largura_linha = 2**3 * 8
memoria_principal = 2**4 * 2**20 * 8

# Quantidade de bits da tag
q_blocos = memoria_principal/largura_linha
largura_tag = int(math.log2(q_blocos))

# Quantidade de linhas
q_linhas = int(total_dados/largura_linha)

# total de bits da cache
memoria_cache = (q_linhas * largura_tag) + total_dados

print(f"Quantidade de bits necessária para a cache = {int(memoria_cache)} bits = {int(memoria_cache/1024)}Kb")
```

# Exemplo 5.6

# Exemplo 5.8

Cálculo da quantidade de bits necessários para uma determinada memória cache, que funciona com mapeamento por conjunto de quatro.

Considere um sistema de computação com uma memória cache de 32KB de capacidade, constituída de linhas com 8 bytes de largura e conjunto de 4. A MP possui uma capacidade de 16MB.

```python
import math

total_dados = 2**5 * 2**10 * 8
largura_linha = 2**3 * 8
memoria_principal = 2**4 * 2**20 * 8

# Quantidade de linhas
q_linhas = int(total_dados/largura_linha)

# Quantidade de bits da tag
q_conjuntos = q_linhas/4
q_blocos = memoria_principal/largura_linha
largura_tag = int(math.log2(q_blocos/q_conjuntos))

# total de bits da cache
memoria_cache = (q_linhas * largura_tag) + total_dados

print(f"Quantidade de bits necessária para a cache = {int(memoria_cache)} bits = {int(memoria_cache/1024)}Kb");
```

# Exemplo 5.9

Cálculo de formato de endereço para memórias cache com mapeamento associativo por conjunto.

Considere uma MP com 64MB de capacidade associada a uma memória cache que funciona com mapeamento associativo por conjunto de 4 e que possui 32KB, com linhas de largura de 16 bytes. Determine o formato do endereço para ser interpretado pelo sistema de controle da cache.

```python
import math

memoria_principal = 2**6 * 2**20
total_dados = 2**5 * 2**10
largura_linha = 2**4

# Quantidade de bits do endereço
bits_endereco = int(math.log2(memoria_principal))
print(f"Endereço: {bits_endereco} bits")

# Quantidade de blocos
q_blocos = memoria_principal/largura_linha

# Quantidaade de linhas
q_linhas = total_dados/largura_linha

# Quantidade de bits do conjunto
q_conjuntos = q_linhas/4
bits_conjunto = int(math.log2(q_conjuntos))
print(f"Conjunto: {bits_conjunto}")

# Quantidade de blocos por conjunto -> tag
blocos_por_conjunto = q_blocos/q_conjuntos
bits_tag = int(math.log2(blocos_por_conjunto))
print(f"Tag: {bits_tag} bits")

# Quantidade de bits da célula
bits_celula = bits_endereco - bits_tag - bits_conjunto
print(f"Célula: {bits_celula} bits")

# Formato do endereço
print(f'''\nEndereço:
Tag           | Conjunto  | Célula
{"0" * bits_tag} | {"0"* bits_conjunto} | {"0" * bits_celula}''')
```

# Exemplo 5.10
Seja uma MP constituída de blocos com largura de 32 bytes, associada a uma cache com 64KB; a cache usa mapeamento por conjunto de 4. Em dado instante o processador realiza um acesso, colocando o seguinte endereço (expresso em algarismos hexadecimais): 3FC92B6. Determine qual deverá ser o valor binário do conjunto que será localizado pelo sistema de controle da cache.

```python
endereco = "3FC92B6"
bits_endereco = 4 * 7
largura_bloco = 2**5
cache = 2**16

# Quantidade de bits da célula
bits_celula = int(math.log2(largura_bloco))
print(f"Célula: {bits_celula} bits")

# Quantidade de bits do conjunto
q_linhas = cache/largura_bloco
q_conjuntos = q_linhas/4
bits_conjunto = int(math.log2(q_conjuntos))
print(f"Conjunto: {bits_conjunto} bits")

# Quantidade de bits da tag
bits_tag = bits_endereco - bits_celula - bits_conjunto
print(f"Tag: {bits_tag}")

# Endereço em binário
endereco_bin = "00" + str(bin(int(endereco, 16)))[2:29]
print(f"Endereço em binário: {endereco_bin}")

# Formato do endereço
print(f'''\nFormato do endereço:
Tag            | Conjunto  | Célula
{endereco_bin[0:14]} | {endereco_bin[14:23]} | {endereco_bin[23:29]}

Conjunto localizado pela cache: {endereco_bin[14:23]}''')
```
